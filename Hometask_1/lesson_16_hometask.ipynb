```python
from nltk.tokenize import word_tokenize
from nltk.stem.porter import PorterStemmer
from gensim import corpora, models
import gensim
from nltk.corpus import stopwords
import re
from nltk.tokenize import RegexpTokenizer
import pandas as pd 
```

```python
fn= 'voted-kaggle-dataset.csv'
df = pd.read_csv(fn)
print ('len of texts= {:,}'.format(len(df)))
index = 10 
print(df.loc[index, 'Description'])
```

```python
datasets_description = df['Description'].dropna()
description_text = datasets_description.values
```

```python
en_stop  = set(stopwords.words('english')).union({'data', 'dataset', 'model', 'content', 'csv', 'contain',})
p_stemmer = PorterStemmer()
    
def tokenize(description_text):
    texts = []
    tokenizer = RegexpTokenizer(r'\b[a-zA-Z]{3,}\b') # just for demo
    
    for desc in description_text:
        # tokenize document string
        raw = desc.lower()
        tokens = tokenizer.tokenize(raw)

        # remove stop words from tokens
        tokens = [token for token in tokens if token not in en_stop]

        # stem tokens
        tokens = [p_stemmer.stem(token) for token in tokens]

        # add tokens to list
        texts.append(' '.join(tokens))
    return texts

texts = tokenize(description_text)
texts[0]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
import pickle
```

```python
vectorizer = CountVectorizer(min_df=10)
vectorizer.fit(texts)
```

```python
description_data_vectorized= vectorizer.transform(texts)
print (description_data_vectorized)
```

```python
corpus = gensim.matutils.Sparse2Corpus(description_data_vectorized, documents_columns=False)
[item for item in corpus][:5]
```

```python
id_map = dict((v, k) for k, v in vectorizer.vocabulary_.items()) 
id_map
```

```python
ldamodel = gensim.models.ldamodel.LdaModel (corpus, num_topics=10, id2word=id_map, passes=25, random_state=34)
```

```python
ldamodel.print_topics(num_topics=10, num_words=15)
```
